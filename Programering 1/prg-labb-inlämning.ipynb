```python
# d

import random as rnd

for x in range(5): 
    x = rnd.uniform(-1,1) # rnd.uniform gör så att den slumpar random mellan bara -1 och 1 
    y= rnd.uniform(-1,1)
    print(x,y) # den slumpar fram 5 gånger ett tal för x mellan -1 och 1, och samma sak för y. Sedan skriver den ut dessa tal.
```

```python
# e

import random as rnd
import math

inne = 0
for x in range(10): 
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)
    print(x,y)
    d = math.sqrt(x**2+y**2) # det här räckna formeln visar om punkten är inne elle utanför.

    if d<=1: # om d(avståndendet från punkten till origo) är lika med eller mindre än 1 så är punkten inom cirkeln
        print("inom cirkel")
        inne = 1+inne #om den är inom cirkeln så ska man adera ett till antalet inne punkter och räckna antalet inne punkter.
    else:
        print("utanför")

andel=inne/10 #sedan delar man antalet inne punkter med 10 och skriva ut svaret

print(andel)


```

```python
# f

import random as rnd
import math

inne = 0
for x in range(10): 
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)
    print(x,y)
    d = math.sqrt(x**2+y**2)

    if d<=1:
        print("inom cirkel")
        inne = 1+inne
    else:
        print("utanför")

andel=inne/10

print(andel*4)

```

```python
# g 

import matplotlib.pyplot as plt
import random as rnd
import math 

inne=0

for x in range(1000): 
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)
    d = math.sqrt(x**2+y**2)
    
    if d<=1:
        plt.plot(x,y,'*r')
        inne=1+inne
    else:
        plt.plot(x,y, '*g')

plt.show()


```

# Uppgift 2 - Burr

```python
# a

for i in range(1,101):
    print(i,end=" ")
```

```python
# b 

for i in range(1,101):
    if i%5 == 0: #om man delar talet med 5 och får en exakt svar, alltså ingen sifra kvar så är tal multipel med 5 och då skriver man ut burr
        print("burr", end=" ")
    else:
        print(i,end=" ")
```

```python
# c

tal= int(input("Ange ett tal"))

for i in range(1,101):
    if i%tal == 0:
        print("burr", end=" ")
    else:
        print(i, end=" ")

```

```python
# d

tal= int(input("Ange ett tal"))
tal2= int(input("Ange även ett tal"))

for i in range(1,101):
    if i%tal == 0:
        print("burr", end=" ")
    elif i%tal2 == 0:
        print("birr", end=" ")
    else:
        print(i, end=" ")
```

```python
start=int(input("Ange ett starttal"))
slut=int(input("Ange ett sluttal"))
tal= int(input("Ange ett tal"))
tal2= int(input("Ange även ett tal"))

for i in range(start,slut):
    if i%tal == 0:
        tal=1+tal
    elif i%tal2 == 0:
        tal2=1+tal2
burr=tal
birr=tal2
burrbirr=tal+tal2

print(f"burr = {burr} ,birr = {birr} och burrbirr = {burrbirr}")
```
