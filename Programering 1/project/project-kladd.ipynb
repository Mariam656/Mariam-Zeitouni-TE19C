```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(2)
intensivvård = [6657, 5771]
död = [3582, 1411]
width = 0.4

plt.xticks(x, ["Män", "Kvinna"])
plt.bar(x-0.2, intensivvård, width, color="#800517")
plt.bar(x+0.2, död, width, color="#B23749")
plt.legend(["död", "intensivvård"])

plt.title("Antal intensivvård och döda enligt kön")
plt.show()
```
