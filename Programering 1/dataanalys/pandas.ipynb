```python
# 1 
import pandas as pd 

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

df
```

```python
# 1a

df_g = df[df["sysselsättning"]=="förvärvsarbetande"]
df_s = df_g["2019"].sum()

print(f"År 2019 var {df_s} personer förvävsarbetande")
```

```python
# 1b

df_e = df[df["sysselsättning"]=="ej förvärvsarbetande"]
df_s = df_e["2019"].sum()

print(f"År 2019 var {df_s} personer ej förvävsarbetande")
```
