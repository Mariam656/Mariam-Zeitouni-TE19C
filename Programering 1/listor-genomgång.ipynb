```python
# skapa en lista 
frukter = ["äppple", "päron", "kiwi", "banan", "jordgubbe", "blåbär"]

#indexering/åtkomstoperator
print(f"Fruktlista: {frukter}")
print(f"frukter[0]: {frukter[0]}")
print(f"frukter[-1]: {frukter[-1]}")
print(f"frukter[3]: {frukter[3]}")
print(f"frukter[::-1]: {frukter[::-1]}")
print("\n")

for frukt in frukter:
    print(f"Frukt:{frukt}")
```

```python
grönsaker = ["tomat", "gurka", "majs", "sallad"]
frukter = ["äppple", "päron", "kiwi", "banan", "jordgubbe", "blåbär"]
fruktsallad = [] # tom lista

# append() - lägger till element i en lista
for grönsak in grönsaker:
    fruktsallad.append(grönsak)

for frukt in frukter:
    fruktsallad.append(frukt)
    
print(fruktsallad)
```

```python
import matplotlib.pyplot as plt

# list comprehension
x = [2*x-2 for x in range(10)]
y = [x for x in range (10)]
print: (f"x:{x}")
print: (f"y:{y}")

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graf till y = 2x-2")
```

```python
# 2D lista
tabell = []

for i in range (11):
    rad = [x*i for x in range(11)] # skapar en rad av gånger taballen
    tabell.append(rad)
    for j in range(11):
        print(f"{tabell[i][j]:3}", end=" ")
    print()

print()

print(tabell)
print(f"tabell[3]: {tabell[3]}")
print(f"tabell[3][3]: {tabell[3][3]}")
```
