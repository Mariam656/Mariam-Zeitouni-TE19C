```python
# 3
import numpy as np 
import matplotlib.pyplot as plt 

x = np.arange(0,11)

def f(k,x,m):
    y = k*x + m
    return y

plt.plot(x,f(3,x,1))
plt.title("Grafen till 3x+1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
